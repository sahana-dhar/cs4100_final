```python
# Extraction with progress
import tarfile
from tqdm.auto import tqdm

# extract from drive and write back to drive (in raw folder)
TAR_PATH    = "/content/drive/MyDrive/OASIS/OAS2_RAW_PART2.tar.gz"
EXTRACT_DIR = "/content/drive/MyDrive/OASIS/raw"

os.makedirs(EXTRACT_DIR, exist_ok=True)

# CLAUDE CODE USED TO UNPACK FILES
with tarfile.open(TAR_PATH, "r:gz") as tar:
    members = tar.getmembers()
    skipped, extracted = 0, 0

    pbar = tqdm(total=len(members), desc="Extracting", unit="file", position=0, leave=True)
    for member in members:
        dest = os.path.join(EXTRACT_DIR, member.name)
        if member.isfile() and os.path.exists(dest):
            skipped += 1
        else:
            tar.extract(member, EXTRACT_DIR)
            extracted += 1
        pbar.set_postfix(extracted=extracted, skipped=skipped)
        pbar.update(1)
    pbar.close()

print(f"\nDone. Extracted: {extracted} | Skipped: {skipped}")
```

# Combine folders into /raw folder

```python
import shutil

PART1_DIR = "/content/drive/MyDrive/OASIS/raw/OAS2_RAW_PART1"
PART2_DIR = "/content/drive/MyDrive/OASIS/raw/OAS2_RAW_PART2"
TARGET_DIR = "/content/drive/MyDrive/OASIS/raw"

for part_dir in [PART1_DIR, PART2_DIR]:
    for subject in os.listdir(part_dir):
        src = os.path.join(part_dir, subject)
        dst = os.path.join(TARGET_DIR, subject)
        if not os.path.exists(dst):
            shutil.move(src, dst)
            print(f"Moved: {subject}")
        else:
            print(f"Skipped (already exists): {subject}")

# Once everything is moved, remove the now-empty wrapper folders
os.rmdir(PART1_DIR)
os.rmdir(PART2_DIR)
print("Done — wrapper folders removed")
```
